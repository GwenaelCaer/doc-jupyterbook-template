Matplotlib is a library for making 2D plots in Python. It is designed with the philosophy that you should be able to create simple plots with just a few commands:

## Quick start

### Initialize

````python
import numpy as np
import matplotlib.pyplot as plt
````

### Prepare

````python
X = np.linspace(0, 10*np.pi, 1_000)
Y = np.sin(X)
````

### Render & Observe

````python
fig, ax = plt.subplots()
ax.plot(X, Y)
plt.show()
````

## Choose your plot

Matplotlib offers several kind of plots

````python
X = np.random.uniform(0, 1, 100)
Y = np.random.uniform(0, 1, 100)

fig, ax = plt.subplots(figsize=(3, 3))
ax.scatter(X, Y)
plt.show()
````

````python
X = np.arange(10)
Y = np.random.uniform(0, 10, 10)

fig, ax = plt.subplots(figsize=(3, 3))
ax.bar(X, Y)
plt.show()
````

````python
Z = np.random.uniform(0, 1, (8, 8))

fig, ax = plt.subplots(figsize=(3, 3))
ax.imshow(Z)
plt.show()
````

````python
Z = np.random.uniform(0, 1, (8, 8))

fig, ax = plt.subplots(figsize=(3, 3))
ax.contourf(Z)
plt.show()
````

````python
Z = np.random.uniform(0, 1, 4)

fig, ax = plt.subplots(figsize=(3, 3))
ax.pie(Z)
plt.show()
````

````python
Z = np.random.uniform(0, 1, 100)

fig, ax = plt.subplots(figsize=(3, 3))
ax.hist(Z)
plt.show()
````

````python
X = np.arange(5)
Y = np.random.uniform(0, 1, 5)

fig, ax = plt.subplots(figsize=(3, 3))
ax.errorbar(X, Y, Y/4)
plt.show()
````

````python
Z = np.random.normal(0, 1, (100, 3))

fig, ax = plt.subplots(figsize=(3, 3))
ax.boxplot(Z)
plt.show()
````

## Tweak your plot

````python
X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(X, Y, color='black')
plt.show()
````

````python
X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(X, Y, linestyle='--')
plt.show()
````

````python
X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(X, Y, linewidth=5)
plt.show()
````

````python
X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(X, Y, marker='o')
plt.show()
````

## Organize your plot

````python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(X, Y1, X, Y2)
plt.show()
````

````python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(3, 3))
ax1.plot(X, Y1, color='C0')
ax2.plot(X, Y2, color='C1')
plt.show()
````

````python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(3, 3))
ax1.plot(Y1, X, color='C0')
ax2.plot(Y2, X, color='C1')
plt.show()
````

## Label your plot

````python
X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(X, Y)
fig.suptitle(None)
ax.set_title('A Sine wave')
plt.show()
````

````python
X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(X, Y)
ax.set_ylabel(None)
ax.set_xlabel('Time')
plt.show()
````
